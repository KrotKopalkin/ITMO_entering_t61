```python
print("VC для категориальных признаков:")
print(transactions_df.vendor_category.value_counts(ascending=True), end="\n\n")
print(transactions_df.vendor_type.value_counts(ascending=True), end="\n\n")
print(transactions_df.vendor.value_counts(ascending=True), end="\n\n")
print(transactions_df.currency.value_counts(ascending=True), end="\n\n")
print(transactions_df.country.value_counts(ascending=True), end="\n\n")
print(transactions_df.city.value_counts(ascending=True), end="\n\n")
print(transactions_df.card_type.value_counts(ascending=True), end="\n\n")
print(transactions_df.is_card_present.value_counts(ascending=True), end="\n\n")
print(transactions_df.channel.value_counts(ascending=True), end="\n\n")
print(transactions_df.is_high_risk_vendor.value_counts(ascending=True), end="\n\n")
print(transactions_df.is_weekend.value_counts(ascending=True), end="\n\n")
print(transactions_df.is_fraud.value_counts(ascending=True), end="\n\n")
```
